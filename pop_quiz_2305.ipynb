Даны N пар чисел. В каждой паре первое число - это целое число от 0 до 99, второе - вещественное от -5 до 5. Задание: составить список длиной 100 элементов, чтобы на индексах, соответствующих первым числам в каждой паре, стояли вторые числа из этих пар. Остальные числа в списке должны получаться из известных с помощью линейной интерполяции. Числа на индексах от 0 до минимального известного и от максимального известного до 99 должны быть равны числам на минимальном известном и максимально известным индексам соответственно.

```python
import random
N = random.randint(5, 15)
pairs = []
X = random.sample(range(3, 97, 3), k=N)
Y = [random.random() * 10 - 5 for _ in range(N)]
pairs = list(zip(X, Y))
print(pairs)
```
