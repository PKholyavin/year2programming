```python
def calc_word_prob(sent, word_1, word_2):
  counter = sent.count(word_1)

  counter_2 = 0
  for index, word in enumerate(sent[:-1]):
      if word == word_1 and sent[index + 1] == word_2:
        counter_2 += 1

  return counter_2 / counter
```

Будем генерировать последовательности слов: выберем случайное слово из словаря и сгенерируем самое вероятное предложение, которое будет с него начинаться.

Можно добавить в текст специальные слова, которые будут обозначать начало и конец текста:

```python
text = "^ a b c a d c a b b c a a d #"
```

```python
text = text.split(" ")
```

```python
vocab = sorted(list(set(text)))
print(vocab)
```

```python
import random
random.choice()
```

```python
word_probs = {}
for i in vocab:
  word_probs[i] = {}
  for j in vocab:
    word_probs[i][j] = calc_word_prob(text, i, j)

print(word_probs)
```
